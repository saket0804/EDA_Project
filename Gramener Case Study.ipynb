### Analyze how term (the loan duration, typically 36 or 60 months) affects default rates.

## Credit History and Utilization

```python
# Understand the distribution of credit utilization

plt.figure(figsize=(10,6))
sns.histplot(Data["revol_util"], bins=30, kde=True, color='blue')
plt.title('Distribution of Credit Utilization (Revolving Utilization Rate)')
plt.xlabel('Revolving Utilization Rate (%)')
plt.ylabel('Frequency')
plt.show()
```

- The bar on the far left shows a high number of borrowers with very low (close to 0%) or zero credit utilization. This indicates that a significant portion of borrowers have not used their available credit at all or have very low revolving balances compared to their credit limits.
- After the initial spike at 0%, the distribution appears relatively uniform across the range from 10% to 80%. This suggests that credit utilization is spread somewhat evenly across borrowers in this middle range.
- Borrowers appear to be using between 10% to 80% of their available credit at a fairly consistent frequency.
- Toward the right of the plot (beyond 80%), there is a noticeable drop-off in the number of borrowers using nearly all of their available credit (i.e., credit utilization approaching 100%). While there are still borrowers in this range, they are fewer in comparison to other ranges.
- Most borrowers either do not use much of their revolving credit (0% utilization) or use it moderately (10-80% range).
- Few borrowers are maxing out their credit (close to 100% utilization), but they do exist.
- Borrowers with very high utilization rates could be more risky, as they are using a large portion of their available credit, which could indicate financial stress.

```python
# Plotting Graph to check default rate in case of credit utilization

plt.figure(figsize=(8,6))
sns.boxplot(x='loan_status', y='revol_util', data=Data)
plt.title('Revolving Utilization by Loan Status')
plt.xlabel('Loan Status')
plt.ylabel('Revolving Utilization (%)')
plt.show()
```

- The median revolving utilization for borrowers with a "Charged Off" status (defaulted loans) is higher than that for those who are "Fully Paid" or "Current."
- Borrowers who defaulted tend to have a higher credit utilization on average, as indicated by the elevated median and the interquartile range for the "Charged Off" group.
